```python
from folium.plugins import HeatMap
import folium
import pandas as pd 
```

```python
df= pd.read_csv('/home/eurekastein/Documentos/PeriodoIV/Rodrigo/nearbyshops/nueva_app/visualization/data.csv')
df.head()
```

```python
df.fecha_de_hechos = pd.to_datetime(df.fecha_de_hechos)
df['month'] = df.fecha_de_hechos.apply(lambda x: x.month)
df['day'] = df.fecha_de_hechos.apply(lambda x: x.day)
df['year'] = df.fecha_de_hechos.apply(lambda x: x.year)
```

```python
def generateBaseMap(default_location=[19.41944, -99.145556], default_zoom_start=12):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map

base_map = generateBaseMap()
base_map
```

```python
df_copy = df[df.month >= 1].copy()
df_copy['count'] = 1
base_map = generateBaseMap()
HeatMap(data=df_copy[['latitud', 'longitud', 'count']].groupby(['latitud', 'longitud']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
base_map
```

```python
df_month_list = []
for month in df_copy.month.sort_values().unique():
    df_month_list.append(df_copy.loc[df_copy.month == month, ['latitud', 'longitud', 'count']].groupby(['latitud', 'longitud']).sum().reset_index().values.tolist())
```

```python
from folium.plugins import HeatMapWithTime
base_map = generateBaseMap(default_zoom_start=11)
HeatMapWithTime(df_month_list, radius=5, gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map
```
